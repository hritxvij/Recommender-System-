### 1h. Finding top movies

We need to recommend only the top n number of movies and not all. For this we have made another function. 

We make a function top_movies with two objects, user_id and top_n. Lets say to consider a movie highly recommended, the ratings has to be good,i.e., the user's rating should match the top_n. So we first find the maximumrating that the user has given. We define a range and say if the max_rating by the person is less than that range, that means the user did not like any movie so far. 

Else, if the user has indeed liked a movie with higher ratings, then what we can do is define list of a 5-star and a 4-star rating movieId. This means that the user will like the 4 or 5-star rated movies.

We can do this usign various filters in the model like;
- if the number of 5-star rated movies are more than top_n number of movies, we randomly return a list of top_n movieId.
- ElseIf, the total numeber of 4 or 5-star rated movies are less than top_n, then we add all the 4 and 5 rated movies.
- otherwise, we return a list of all 5-rated movies and a few random selection from 4-rated movies.



```python
def top_movies (user_id, top_n=5):
    
    ## making a threshold to like a movie based on rating
    max_rating = max(ratings.loc[(ratings.userId==user_id), "rating"])
    
    ## in our case, threshold rating is 4
    if max_rating < 4:
        print ("None liked so far!")
        return (None)
    
    
    else:
        
        ## condition for 5-rated movies
        movie_5 = ratings.loc[(ratings.userId==user_id)&(ratings.rating==5), "movieId"]
        
        ## condition for 4-rated movies
        movie_4 = ratings.loc[(ratings.userId==user_id)&(ratings.rating==4), "movieId"]
        
        ## If number of 5-rated movies are more than top_n, then return a List of randomly selected movie
        if len(movie_5) > top_n:
            return(list(np.random.choice(movie_5,5)))
        
        
        ## If the total number of 5-rated movies & 4-rated movies are less than top_n, then return a List
        elif len(movie_5) + len(movie_4) < top_n:
            return(list(movie_5)+list(movie_4))
        
        
        else:
            ## else return a list containing all the 5-rated movies and a few random selection from the 4 rated movies
            return(list(movie_5) + list(np.random.choice(movie_4, top_n-len(movie_5))))
```

Here, we make a function to get similar movies. Firstly, we give two objects, user_id and sim_score. Based on the user_id called, we get it corresponding title. Next, using the similarity matrix we found, we find the row where the title belongs. For which, we then find the scores of the movies that are similar. Hence, when we call the user_id and the range of similarity score, we get a list of similar movie.

**Keep in mind:** 

*A similarity score closer to 1 is more similar than the one away from it.*

```python
def get_similar_movies (movie_id, sim=0.9):
    
    ## getting the title for the corresponding movie id
    title = movie_list.loc[movie_list.movieId==movie_id, "title"]
    title = np.array(title)[0]
    
   
    ## getting the row containing that movieID from the cosine similarity matrix
    similarity = cos_movies.loc[title,:]
    
    
    ## similarity scores of all movies
    scores = np.array(similarity)
    
    
    ## related movies that had similarity score of more than sim_value
    related = cos_movies.loc[title,: ][cos_movies.loc[title,:]>sim].index
    
    
    return(list(related))
```

```python
## function definition
def recommender(movie_name, data,model, n_recommendations ):
    ## fitting the data inside the model
    model.fit(data)
    
    ## indexing the movie in the dataset that match the movie_name
    ## 2 to choose the second column
    idx=process.extractOne(movie_name, movies['title'])[2]
        
    
    print('Searching for recommendations.....')
    
    ## calculating distance of similar movies using NN based on index of the movie extracted
    ## then the indexes of top rated movie's  are stored in indices
    distances, indices=model.kneighbors(data[idx], n_neighbors=n_recommendations)
    
    ## for every movie in indices
    for i in indices:
        
        ## printing the movie from the datset that corresponds to the index i
        ## but we avoid those movies which are same or have same index
        print(movies['title'][i].where(i!=idx))

```
